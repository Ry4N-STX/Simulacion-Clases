# Concurrencia Lineal


```python
from pandas import DataFrame
xn=0
ui=0
xn1=[]
result=[]
aux=0
#datos de ingreso
print("Ingrese los Datos")
print("Variable A : ")
valorA=int(input())
print("Variable C : ")
valorC=int(input())
print("Variable M : ")
valorM=int(input())

variableA=valorA
variableC=valorC
variableM=valorM

print("Semilla : ")
valorsemilla=int(input())
semilla=valorsemilla

print("Cantidad de Itraciones: ")
valoriteraiones=int(input())
iteraciones=valoriteraiones

xn=semilla
#proceso de concurrencia lineal
for i in range(0,iteraciones):
    aux=((variableA*xn)+variableC)%(variableM)
    ui=aux/variableM
    xn=aux
    xn1.append(xn)
    result.append(ui)

    
#tabla de resultados
tabla_result={'Xn':xn1,'Ui':result}
dataframe=DataFrame(tabla_result)
dataframe
```
