# METODO CUADRADOS MEDIOS

```python
from tabulate import tabulate
from pandas import DataFrame
print("Escriba la semilla :")
semt=int(input())
print("Escriba la cantidad de digitos :")
cantidadD=int(input())
print("Escriba la cantidad de iteracipnes :")
iteraciones=int(input())
semilla=semt
cantidad=iteraciones
semillacuadrada=0
digitos=cantidadD
scuadrado=[]
datosXN=[]
datosXN2=[]
rndatos=[]
tamanio=[]

#print("SEMILLA DE :",semilla)
for i in range(0,cantidad):
    semillacuadrada=semilla*semilla
    datosXN.append(semilla)
    xn=str(semillacuadrada)
    longit=len(xn)
    tamanio.append(longit)
    d1=int((int((longit/2))-int((digitos/2))))
    d2=int((int((longit/2))+int((digitos/2))))
    semillanueva=xn[int(d1):int(d2)]
    nuevo=int(semillanueva)
    inicio="1"
    for j in range(0,digitos):
        inicio=inicio+"0"
    RN=(nuevo/int(inicio))
    semilla=nuevo
    datosXN2.append(nuevo)
    scuadrado.append(semillacuadrada)
    rndatos.append(RN)

#print("DIGITOS :",cantidadD)

dict_listas = {'XN' : datosXN, 'XN*XN' : scuadrado,'LONGITUD' : tamanio,'UI' : datosXN2,'RN' : rndatos,}
dataframe = DataFrame(dict_listas)
dataframe
```
