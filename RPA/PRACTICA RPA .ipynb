# PDF
# Unir PDF

```python
# Al igual que en ejemplo anterior usaremos dos pdf diferentes de internte que se descargaran
# para ser unidos puede ser la cantidad que quieran.
example_pdf = download_file_from_url('http://www.webgou.uma.es/archivos/comunicacion.pdf')
example_pdf_2 = download_file_from_url('https://www.uv.es/bellochc/pdf/pwtic2.pdf')

# Unie el archivo PDF 2 veces con otro o los que se desee, 
# también podría unirse el pdf asimismo las veces que quiera
merged_pdf = join_pdf_files([example_pdf, example_pdf_2])

# Abrimos el resultado de la union, este se mostrara en una nueva pestaña y a la vez se mostrara la ruta donde 
# se a guardado
open_file(merged_pdf)
```

# PDF
# Extraer pagina de PDF
Extrae un rango particular de un PDF a un archivo separado.

```python
# De igual forma usaremos un pdf de internet en eset caso uno que trata sobre la Historia de la computacion como ejemplo
example_pdf = download_file_from_url('http://atc2.aut.uah.es/~frutos/fundamentos/pdf/Historia.pdf')

# Extreremos del PDF desde la pagina 4 a la 7
new_file = extract_page_range_from_pdf(example_pdf, 4, 7 )
# Abrirmos el archivo para apreciar la extraccion
open_file(new_file)
# tambien se muestra la ruta donde se almacena el nuevo archivo
```
